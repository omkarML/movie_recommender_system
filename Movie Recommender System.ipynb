### Item based collaborative filter on entire data set

```python
matrix_df
```

```python
movie_correlation = matrix_df.corr(method = 'pearson' , min_periods = 80)
```

```python
movie_correlation
```

```python
my_rating = pd.read_csv(r'C:\Users\HP\Desktop\DS\udemy\DL and ML Practical Tutorials - Package\Project 8\My_Ratings.csv')
```

```python
my_rating
```

```python
my_rating['Movie Name'][0]
```

```python
same_movie_list = pd.Series()

for i in range(0,2):
    same_movie = movie_correlation[ my_rating['Movie Name'][i]  ].dropna()
    same_movie = same_movie.map(lambda x: x* my_rating['Ratings'][i])
    same_movie_list = same_movie_list.append(same_movie)
```

```python
same_movie_list.sort_values(inplace = True , ascending = False)
print(same_movie_list)
```
